### Data Preprocessing

#### Convert Attrition from ('Yes', 'No') to (1,0) 

```python
def convert(x):
    if x=='Yes':
        return 1
    else:
        return 0
```

```python
df['Attrition']=df['Attrition'].apply(convert)
```

```python
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
newdf=pd.get_dummies(df,drop_first=True)
```

```python
newdf.head()
```

#### Check the shape of our new dataset

```python
newdf.shape
```

#### Print unique values in our new dataframe

```python
for column in newdf.columns:
    print(f'Unique values in {column} are : {newdf[column].unique()}')
```

#### Split the columns into input and target variables (X and y)

```python
X=newdf.drop(columns=['Attrition'])
y=newdf['Attrition']
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
X.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print(f'Accuracy score of our model is : {metrics.accuracy_score(y_test,y_pred)}')
print('Confusion matrix is :')
print(f'{metrics.confusion_matrix(y_test,y_pred)}')
print('Classification report is:')
print(metrics.classification_report(y_test,y_pred))
print(f'cross Value score is : {cross_val_score(model,X,y).mean()}')
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
sns.heatmap(metrics.confusion_matrix(y_test,y_pred),annot=True,fmt='d')
plt.show()
```

### Random Forest Classifier
** Choose the best estimator and parameters :GridSearchCV**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
forest_params=[{'max_depth':[0.5,1,5,10],'n_estimators':[16,32,50,100],'n_jobs':[1,2],'random_state':[1,10,20,42]}]
```

```python
clf=GridSearchCV(RandomForestClassifier(),forest_params,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
print(f'Best estimator is : {clf.best_estimator_}')
print(f'Best parameters are : {clf.best_params_}')
print(f'Best score is : {clf.best_score_}')
```

<span style = "color:blue"> Create Random forest model with the best parameters </span>

```python
model=RandomForestClassifier(max_depth=10, n_estimators=32, n_jobs=1, random_state=20)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python

print(f'Accuracy score is : {metrics.accuracy_score(y_test,y_pred)}')
print('Confusion matrix is :')
print(f'{metrics.confusion_matrix(y_test,y_pred)}')
print('Classification report is:')
print(metrics.classification_report(y_test,y_pred))
print(f'cross Value score is : {cross_val_score(model,X,y).mean()}')
```

<span style = "color:orange"> Visualize confusion matrix </span>

```python
sns.heatmap(metrics.confusion_matrix(y_test,y_pred),annot=True,fmt='d')
plt.show()
```

## Use PCA to reduce dimensionality of the data

#### Import PCA and fit our X_train

```python

from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=0.95)
```

```python
pca.fit(X_train)
```

#### Apply the mapping (transform) to both the training set and the test set.

```python
train_X = pca.transform(X_train)
test_X = pca.transform(X_test)
```

###Import machine learning model of our choice, I am going with RandomForest for this problem

```python
from sklearn.ensemble import RandomForestClassifier
```

#### Create RandomForest model with the best parameter we got earlier and train it

```python
rfc=RandomForestClassifier(max_depth=10, n_estimators=32, n_jobs=1, random_state=20)
```

```python
rfc.fit(train_X,y_train)
```

#### Check the score of our model

```python
rfc.score(train_X,y_train)
```

#### Make predictions with X_test and check the accuracy score

```python
y_pred=rfc.predict(test_X)
```

```python
metrics.accuracy_score(y_test,y_pred)
```

### Print Confusion matrix and Classification report
